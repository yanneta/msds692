Navigate to the official USF website in Chrome: "https://www.usfca.edu/".

```python
website = "https://www.usfca.edu/"
# Create a WebDriver instance for Chrome
driver = webdriver.Chrome()
driver.maximize_window()
# Visit the website
driver.get(website)
```
